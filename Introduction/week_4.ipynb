```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = {
    'Employees_status': ['Full-time', 'Part-time', 'Full-time', 'Part-time', 'Full-time'],
    'Amount_spent': [200, 150, 300, 120, 250]
}
df = pd.DataFrame(data)

# # Suppress warnings
#np.warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning)

# Create the boxplot
fig, ax = plt.subplots(figsize=(6, 6))
df.boxplot(by='Employees_status', column=['Amount_spent'], ax=ax, grid=False)
plt.title('Amount Spent by Employee Status')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Employee Status')
plt.ylabel('Amount Spent')
plt.show()
```
